```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
av_lego=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyExpCon.npy")
av_hard=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyLegExp.npy")
std_lego=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyExpLegStd.npy")
std_hard=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyExpConStd.npy")

TCon=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyDataConT.npy")
TLeg=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyDataLegT.npy")

std_con=np.std(TCon,axis=1)
avg_con=np.average(TCon,axis=1)

std_leg=np.std(TLeg,axis=1)
avg_leg=np.average(TLeg,axis=1)

resolution=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyRes.npy")
#resolution=np.load("C:/Users/dexte/github/RoboSkin/Code/Models/surfaceModel/accuracyResTest.npy")
avg_res=np.average(resolution,axis=1)
std_res=np.std(resolution,axis=1)
```

```python
plt.plot(av_hard,label="Hard surface model")
plt.plot(av_lego,label="Lego surface model")
plt.title("Accuracy plots of different model classification")
std2=std_hard/2
std2[std2>1]=1
std1=std_lego/2
std1[std1>1]=1
plt.errorbar(range(len(av_hard)), av_hard, yerr=std2, errorevery=10)
plt.errorbar(range(len(av_lego)), av_lego, yerr=std1, errorevery=10)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(loc="lower right")
plt.show()
```

```python
#colours=["snow","rosybrown","lightcoral","indianred","brown","firebrick","maroon","darkred","red"]
for i in range(1,len(avg_con)):
    #plt.plot(avg_con[i],c=(0.1+i/10, 0.1, 0.1),label="Hard surface model T="+str(i+1))
    plt.plot(avg_leg[i],c=(0.1, 0.1, 0.1+i/10),label="Lego surface model T="+str(i+1))
    plt.title("Accuracy plots of different T sizes")
    std2=std_con[i]/2
    std2[std2>1]=1
    std1=std_leg[i]/2
    std1[std1>1]=1
    #plt.errorbar(range(len(avg_con[i])), avg_con[i], c=(0.1+i/10, 0.1, 0.1), yerr=std2, errorevery=10)
    plt.errorbar(range(len(avg_leg[i])), avg_leg[i], c=(0.1, 0.1, 0.1+i/10), yerr=std1, errorevery=10)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(loc="lower right")
plt.show()
```
